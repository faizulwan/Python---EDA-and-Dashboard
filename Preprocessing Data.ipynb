## Aggregat Function - Data Frame
- Menggunakan salah satu kolom bertipe data numerik
- Kolom tersebut akan dikombinasikan dengan kolom kategorik

## Syarat 
- Parameter values HARUS DIISI ==> Berisi kolom numerik 
- Parameter 'aggfunc' HARUS DIISI ==> Berisi aggfunct yang akan digunakan

```python
df.head()
```

```python
pd.crosstab(index=df['sub_category'], columns='harga rata-rata', values=df['unit_price'], aggfunc='mean')
```

```python
pd.crosstab(index=df['format'], columns=df['sub_category'], values=df['unit_price'], aggfunc='max').round(2)
```

```python
pd.crosstab(index=df['format'], columns=df['sub_category'], values=df['unit_price'], aggfunc='min').round(2)
```

```python
## Perbedaan Crosstab dan Pivot
- Pada crosstab tidak dapat memfilter sedangkan pivot dapat
- Pada pivot table, fungsi aggregate DEFAULTNYA adalah MEAN
- Pada crosstab fungsi agg nya adalah COUNT
```
