### 9. Tampilkan kategori film yang memiliki rata-rata durasi (_average of length_) di atas rata-rata kategori film lainnya!

```python
sql_df('SELECT AVG(length) FROM film_list')
```

```python
sql_df('SELECT category, AVG(length) AS ave FROM film_list GROUP BY category HAVING ave > (SELECT AVG(length) FROM film_list) ORDER BY ave DESC')
```
