```python
## Handling Missing Value
Opsi handling Missing Value ==> DROP atau FILL (Data sintetis)

- DROP :
    - Kolom
    - Baris
    
DROP Kolom :
    - Terdapat Kolom Pengganti : Misal Missing Value ada di kolom Jarak (m), Kita memiliki Kolom jarak (km)
                                - Missing Value di Gaji (Rp), Kita memiliki Gaji (USD)
    - 70% - 80% Missing Value nya
DROP Baris :
    - Lihat berapa banyak Kolom yg kosong di baris tsb
    - Pastikan Jumlah Datanya Banyak
    
Misal kita ingin prediksi : MACET : Ngga
        Macet : 990
        Ngga : 10
Features & Target ==> Machine Learning

Jika Ada Missing Value pada Kolom Target ==> Langsung DROP BARIS

FILLNA : Mengisi Data Sintetis
    - Harus melihat Kolom / Features lain 
    - Mean ===> Data Numerik dan Tidak ada Outliers
    - Median ==> Data Numerik
    - Modus ==> Data Kategorikal
    
Kolom Gaji ==> Missing Value
- Kota / Lokasi
- Jabatan
df [df['jabatan'] == 'manajer' & df['lokasi'] == 'solo'].mean()
```

```python
### Pengecekan Outliers
Outliers adl Data di atas Upper Fence atau di bawah Lower Fence

Upper fence - Batas Atas = Q3 + (1.5 * IQR)
Lower Fence - Batas Bawah = Q1 - (1.5 * IQR)

IQR = Q3 - Q1

Q3 = Kuartil 3 == Titik tengah antara nilai Median dan nilai Max
Q1 = Kuartil 1 == Titik tengah antara nilai Median dan nilai Min
```
