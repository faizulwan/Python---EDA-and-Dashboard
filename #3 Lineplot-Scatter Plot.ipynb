## 1. Line Plot

```python
# Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Flights Dataset from Seaborn
flights = sns.load_dataset("flights")
flights.head(3)
```

### 1.3 Lineplot using Matplotlib

```python
plt.figure(figsize=(8,6))                                     # Figure Size
plt.plot(flights['year'][flights['month']=='Jan'],            # X axis = year name only in January
        flights['passengers'][flights['month']=='Jan'],       # Y axis = number of passengers only in January
        'blue')                                                # Line color
plt.title('Number of Passenger in January Yeearly', size=17)  # Title
plt.xlabel('Yearly')                                          # X Label
plt.ylabel('Number of Passengers')                            # Y Label
plt.show()
```

### 1.2 Lineplot Using Seaborn

```python
sns.lineplot(data=flights, x='year', y='passengers', color='red')                 # Create line plot using seaborn
plt.title('Number of Passengers-Year', size=17)                                   # Title
plt.xlabel('Yearly')                                                              # X Label
plt.ylabel('Number of Passengers')                                                # Y Label
plt.grid(True)
plt.show()
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Number of Passengers-Year', size=17)                                   # Title
plt.xlabel('Yearly')                                                              # X Label
plt.ylabel('Number of Passengers')                                                # Y Label
plt.grid(True)
plt.show()
```

### 1.3 Lineplot Using Pandas

```python
df_january = flights[flights['month']=='Jan']                # Filtering passengers in January
df_january.plot.line(x='year', y='passengers', legend=False) # Cretae line plot using Pandas
plt.title('Number of Passengers in January', size = 20)      # Title
plt.show()
```

## 2. Scatter Plot

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

### 2.1 Scatter using Matplotlib

```python
plt.figure(figsize=(8,6))
plt.scatter(tips['total_bill'],tips['tip'])
plt.title('Total Bill - Tip Scatter Plot', size=18)
plt.xlabel('Total Bill')                                                              # X Label
plt.ylabel('Tip')   
plt.show
```

### 2.2 Using Seaborn

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', legend=True)
plt.title('Total Bill - Tip Scatter Plot', size=18)
plt.xlabel('Total Bill')                                                              # X Label
plt.ylabel('Tip')
plt.show()
```

### 2.3 Using Pandas

```python
tips.plot.scatter(x='total_bill', y='tip')               
plt.title('Total Bill - Tip Scatter Plot', size=18)    # Title
plt.xlabel('Total Bill')                               # X Label
plt.ylabel('Tip')                                      # Y Label
plt.show()
```

```python
tips_male = tips[tips['sex']=='Male']
tips_male.head()

tips_female = tips[tips['sex']=='Female']
tips_male.head()

tips_male.plot.scatter(x='total_bill', y='tip', color='red')   
tips_female.plot.scatter(x='total_bill', y='tip')  
plt.show()
```

## 3. Bar Plot

### 3.1 Matplotlib

```python
# Import Dataset
tips = sns.load_dataset("tips")

tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day.head(5)
```

```python
x = tips['day'].unique().tolist()  #Fungsi unique pandas mengubahkan ke list
x
```

```python
# Create a Bar Plot
x = tips['day'].unique().tolist()    #Fungsi unique pandas
y = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.bar(x, y, color = ['green', 'blue', 'red', 'yellow'])
plt.title('Barplot: Average Tips Everyday', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.show()
```

### 3.2 Seaborn

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Barplot: Average Tips Everyday by Gender', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.legend(loc='upper left')
plt.show()
```

```python
sns.barplot(data=tips, x='tip', y='day', hue='sex')
plt.title('Barplot: Average Tips Everyday by Gender', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.legend(loc='center right')
plt.show()
```

### 3.3 Pandas

```python
tips_by_day = tips[['tip','day']].groupby(['day'], as_index=False).mean()  #loc() dan iloc()
tips_by_day 
```

```python
tips_by_day.plot.bar(x='day', y='tip', legend=False)
plt.title('Barplot: Average Tips Everyday by Gender', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.show()
```
